````python
import math

def parabola(a, b, c):
    """
    Solve the equation:
    a x^2 + b x + c = 0
    """
    # Compute the determinant
    d = b**2 - 4*a*c
    
    # Cases
    if d > 0:
        # Two distintc roots
        d2 = math.sqrt(d)
        x1 = (-b+d2)/(2*a)
        x2 = (-b-d2)/(2*a)
        roots = [x1,x2]
    elif d < 0:
        # No real solutions
        roots = [None, None]
    else:
        # One double root
        x1 = -b/(2*a)
        roots = [x1, x1]
    return roots
````

````python
parabola(1,-4,-5)
````
